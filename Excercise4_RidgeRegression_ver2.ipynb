```python
import pdb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import time
import sklearn
#import statsmodels.formula.api as sm
import random
```

```python
mu,sigma=0.0,1.0
def create_random(num):
    pd_rondom=np.random.normal(mu, sigma, num)
    pd_rondom=pd.DataFrame(pd_rondom)
    return pd_rondom
```

```python
def make_x(element_num,data_num):
    x_matrix=create_random(element_num)
    for i in range(data_num-1):
        x_small=create_random(element_num) #train_w = pd.DataFrame(fixed_random)[0:100]
        x_matrix = pd.concat([x_matrix,x_small],axis=1)#pd.concat([create_random(100),create_random(100)],axis=1)
    return x_matrix
```

```python
# return y 
def calc_y(w,X,n):
    a=pd.DataFrame(X.dot(w))
    b=pd.DataFrame(n)
    calcud_y = pd.DataFrame(np.array(a) + np.array(b))
    return calcud_y
```

```python
# This is the unique function of RidgeRegression
# For better speed, do not calculate inverse
def w_estimate(X,y,r_param):
    X = np.matrix(X)
    y = np.matrix(y)
    x_tranposed = X.T
    x_squared =np.matrix(x_tranposed.dot(X))
    x_squared_length = len(x_squared)
    Iden = np.matrix(np.identity(x_squared_length))
    A = x_squared + (r_param*len(X)*Iden)
    b = (x_tranposed).dot(y) 
    w_estimate = np.linalg.solve( A, b )
    
    return pd.DataFrame(w_estimate)
```

```python
# Returns 10^-6 ~ 10^3
def make_r_pram():
    arr_6_3 = np.linspace(-6,3,10)
    arr_power10 = 10**arr_6_3
    return arr_power10
```

```python
def mean_squared_error2(X,w,y):
    l=len(X)*1.0
    term1 = np.matrix(X.dot(w))- np.matrix(y)
    term1=term1.T
    term2 = np.matrix(X.dot(w)) - np.matrix(y)
    te = term1.dot(term2)
    if(te<0):print "mse is minus"
    return te/l
```

```python

traindata_num=100
testdata_num=500
num_of_element=10

# Returns mse for both train and test

def part_a_b_c(traindata_num,testdata_num,num_of_element,r_param):
    w = create_random(num_of_element) # traindata_num x 1
    n_train = create_random(traindata_num) # ytrain_num x 1
    n_test = create_random(testdata_num)
    x_train = (make_x(num_of_element,traindata_num)).T  #100x5, train_num x element
    x_test  = (make_x(num_of_element,testdata_num)) .T
    y_train = calc_y(w,x_train,n_train)
    y_test  = calc_y(w,x_test,n_test)
    w_estimated=w_estimate(x_train,y_train,r_param)
    
    mse_test= mean_squared_error2(x_test,w_estimated,y_test)[0,0]
    mse_train= mean_squared_error2(x_train,w_estimated,y_train)[0,0]
    
    return [mse_train,mse_test]
```

```python

#### a ####
repeat_num =1
r_the_pwr_of_10 = make_r_pram()
train_array=[]
test_array=[]
for r in r_the_pwr_of_10:
    traindata_num=100
    testdata_num=500
    num_of_element=10
    tr100_el10_train = np.zeros(repeat_num)
    tr100_el10_test = np.zeros(repeat_num)
    for i in range(repeat_num):
        trin_tst = part_a_b_c(traindata_num,testdata_num,num_of_element,r)
        tr100_el10_train[i] = trin_tst[0]
        tr100_el10_test[i] = trin_tst[1]
    
    train_array.append(tr100_el10_train.mean())
    test_array.append(tr100_el10_test.mean())
    

r_the_pwr_of_10 = make_r_pram()
train_array10=[]
test_array10=[]
for r in r_the_pwr_of_10:
    #print r
    traindata_num=10
    testdata_num=500
    num_of_element=10
    tr10_el10_train = np.zeros(repeat_num)
    tr10_el10_test = np.zeros(repeat_num)
    for i in range(repeat_num):
        trin_tst = part_a_b_c(traindata_num,testdata_num,num_of_element,r)
        tr10_el10_train[i] = trin_tst[0]
        tr10_el10_test[i] = trin_tst[1]
    
    train_array10.append(tr10_el10_train.mean())
    test_array10.append(tr10_el10_test.mean())
    
    #print "mean tr100_el1_train", tr100_el10_train.mean()
    #print "mean tr100_el1_test", tr100_el10_test.mean()
    
    
```
